```python
# Sales Data Analysis using Python

## Project Description 
This project focuses on analyzing a retail sales dataset to understand sales performance across different product categories, markets, and time periods.  
The objective is to derive meaningful business insights using data analysis and visualization techniques.

## Tools Used
- Python
- Pandas
- Matplotlib
```

```python
## 1. Importing Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## 2. Loading the Dataset
df = pd.read_csv("Superstore.csv")
df.head()
```

```python
# checking number of rows and columns
df.shape
```

```python
# checking column names, data types and null values
df.info()
```

```python
# showing basic statistics like mean, min, max of numeric columns
df.describe()
```

```python
# checking missing values in each column
df.isnull().sum()
```

```python
# removing duplicate rows from the dataset
df = df.drop_duplicates()
```

```python
# converting Order.Date column into datetime format
df['Order.Date'] = pd.to_datetime(df['Order.Date'], errors='coerce')
```

```python
# checking converted Order.Date values
df['Order.Date'].head()
```

```python

category_sales = df.groupby('Category')['Sales'].sum()
category_sales
```

```python
# plotting bar chart for sales by category
category_sales.plot(kind='bar', title='Sales by Category')
```

```python

market_sales = df.groupby('Market2')['Sales'].sum()
market_sales.plot(kind='bar', title='Sales by Market')
```

```python

yearly_sales = df.groupby('Year')['Sales'].sum()
yearly_sales.plot(kind='line', title='Yearly Sales Trend')
```

```python

top_subcat = df.groupby('Sub.Category')['Sales'].sum().sort_values(ascending=False).head(10)
top_subcat.plot(kind='bar', title='Top 10 Sub-Categories by Sales')
```
